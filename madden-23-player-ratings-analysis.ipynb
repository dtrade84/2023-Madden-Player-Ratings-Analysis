# Cleaning the Data
***

In this section we will clean the dataset.

We will be removing unecessary columns, and renaming them.

We will also search for duplicates and remove those as well.

<a id='removecol'></a>
## Removing Columns

First, we will remove any columns we do not need.

We will acheive this using the drop function.

```python
df.head()
```

As we can see there is a lot of columns we do not need. 

Lets get those out of the way.

```python
df.drop(columns=['plyrAssetname', 'plyrBirthdate', 'status', 'iteration', 'plyrPortrait', 'fullNameForSearch', 'primaryKey', 'Position'], inplace=True)
```

```python
df.head()
```

Now we only have the necessary columns.

Onto renaming them.

<a id='renamecol'></a>
## Renaming the Columns

I noticed that all the column names have an undescore and then rating at the end. i will be removing this  portion of the string.

```python
df.columns = df.columns.str.replace(r'_rating$', '')
```

```python
df.head()
```

Next, we search for duplicates.

<a id='duplicates'></a>
## Searching for Duplicates

```python
# Searching for duplicates

sum(df.duplicated())
```

As we can see there are no duplicates.
